Take as given a multi-dimensional Black-Scholes model with $d$ assets
(stocks), whose prices under the risk-neutral probability $Q$ is given
by
$$S^i_t = S^i_0e^{\left(r - \frac{1}{2}\sigma_i^2\right)t + \sigma_i W^{iQ}_t}$$
for all $t\ge0$ and $i=1,\ldots,d$. Here $W^Q=(W^{1Q},\ldots,W^{dQ})$ is
a $d$-dimensional Brownian motion with correlation matrix $C$, in other
words, $$ \mathop{\text{cov}} \left(W^Q_T\right) = C T.$$

<span class="theorem-title">**Exercise 1**</span> Design a procedure
that samples $d$ independent random variables
$Z_1,\ldots,Z_d\sim N(0,1)$ repeatedly and uses it to produce $n$
simulations of $S^1_T, \ldots, S^d_T$.

This is a pen-and-paper exercise, no coding required. Your procedure
should mention the factorization of the matrix $\Sigma$, but there is no
need to go into any detail on the Cholesky decomposition (for example).

<span class="theorem-title">**Exercise 2**</span> Design a procedure
that samples $d$ independent random variables
$Z_1,\ldots,Z_d\sim N(0,1)$ repeatedly and uses it to produce $n$
simulations of $S^1_T, \ldots, S^d_T$.

Then write a function to perform a plain Monte Carlo estimate with a
given number $n$ of simulations of the price at time $0$ of an exchange
option with maturity $T$ and payoff
$$X = \left[\tfrac{1}{3}\sum_{i=2}^4 S^i_T - S^1_T\right]_+$$ in a
multidimensional Black-Scholes model.

Report the estimate and standard error for the price of the option,
given $n=100000$, $T=1$, $d=4$, $S^1_0 = S^2_0 = S^3_0 = 50$,
$S^4_0 = 45$, $r=0.03$, $\sigma_1=\sigma_2=0.2$, $\sigma_3=\sigma_4=0.3$
and $$ \Sigma = \begin{bmatrix}
\phantom-1.0 & \phantom-0.1 & -0.2 & \phantom-0.2 \\
\phantom-0.1 & \phantom-1.0 & \phantom-0.3 & -0.1 \\
-0.2 & \phantom-0.3 & \phantom-1.0 & \phantom-0.5 \\
\phantom-0.2 & -0.1 & \phantom-0.5 & \phantom-1.0
\end{bmatrix}.$$
